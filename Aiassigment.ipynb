```python
##### Question 7
computer = int(input("Enter marks of subject 1: "))
english = int(input("Enter marks of subject 2: "))
math = int(input("Enter marks of subject 3: "))
urdu = int(input("Enter marks of subject 4: "))
science = int(input("Enter marks of subject 5: "))

total = computer + english + math + urdu + science
percentage = total / 5

print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 80:
    print("Grade: A+")
elif percentage >= 70:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```
