```python
#Assigment 4
#Question 1
#Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)

if num2 != 0:
    print("Division:", num1 / num2)
else:
    print("Division not possible")

print("Power:", num1 ** num2)
```

```python
#Question 2
#numeric value using for loop

my_list = ["apple", "banana", 10, "orange"]

for item in my_list:
    if isinstance(item, (int, float)):
        print("Numeric value found:", item)
        break
else:
    print("No numeric value found")
```

```python
#Question 3
#Add a key to a dictionary
my_dict = {"name": "shehzar", "age": 24}

my_dict["city"] = "jamshoro"

print(my_dict)
```

```python
#Question 4
#Sum all numeric items in a dictionary
my_dict = {"a": 10, "b": 20, "c": "hello", "d": 5}

total = 0
for value in my_dict.values():
    if isinstance(value, (int, float)):
        total += value

print("Sum:", total)
```

```python
#Question 5
#Identify duplicate values from a list
my_list = [1, 2, 3, 2, 4, 1, 5]

duplicates = []

for item in my_list:
    if my_list.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Duplicate values:", duplicates)
```
